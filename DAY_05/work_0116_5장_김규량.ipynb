(1) 데이터셋과 메모리 최적화

```python
import pandas as pd
```

```python
file='../DATA/employees.csv'
```

```python
pd.read_csv(file)
```

```python
pd.read_csv(file, parse_dates = ["Start Date"]).head()
```

```python
employees = pd.read_csv(file, parse_dates = ["Start Date"])
```

```python
employees.info()
```

```python
employees["Mgmt"].astype(bool)
```

```python
employees["Mgmt"] = employees["Mgmt"].astype(bool)
```

```python
employees.tail()
```

```python
employees.info()
```

```python
employees["Salary"].fillna(0).tail()
```

```python
employees["Salary"].fillna(0).astype(int).tail()
```

```python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
```

```python
employees.nunique()
```

```python
employees["Gender"].astype("category")
```

```python
employees["Gender"] = employees["Gender"].astype("category")
```

```python
employees.info()
```

```python
employees["Team"] = employees["Team"].astype("category")
```

```python
employees.info()
```
